# Breast Cancer Classification: Data Exploration, Preprocessing, and Model Evaluation


### Dataset Description

This dataset contains information on female breast cancer patients diagnosed between 2006 and 2010 with infiltrating duct and lobular carcinoma. The data was obtained from the 2017 November update of the SEER (Surveillance, Epidemiology, and End Results) Program of the National Cancer Institute (NCI), which provides population-based cancer statistics.

The dataset includes 4,024 patients, focusing on those with known tumor sizes, examined regional lymph nodes, positive regional lymph nodes, and survival months greater than 1. Patients with missing values or survival months less than 1 were excluded.

The columns in the dataset provide details about the patients' demographics, cancer stage, tumor characteristics, lymph node examination results, hormone receptor status, survival status, and survival months.

Key features:
- **Age**: Age of the patient.
- **Race**: Ethnicity of the patient.
- **Marital Status**: Marital status at diagnosis.
- **T Stage**: Tumor size classification.
- **N Stage**: Regional lymph node involvement.
- **6th Stage**: Overall cancer stage based on the 6th edition of the AJCC Cancer Staging Manual.
- **Differentiation**: Tumor cell differentiation (grade).
- **Grade**: Tumor grade, indicating aggressiveness.
- **A Stage**: Anatomical stage of cancer.
- **Tumor Size**: Size of the tumor.
- **Estrogen Status**: Estrogen receptor status (positive or negative).
- **Progesterone Status**: Progesterone receptor status (positive or negative).
- **Regional Node Examined**: Number of lymph nodes examined.
- **Regional Node Positive**: Number of lymph nodes that tested positive.
- **Survival Months**: Number of months the patient survived.
- **Status**: Whether the patient was alive or dead at the time of follow-up.

### Modeling 

```python
# Split the dataset into training and testing sets
x=df_cancer.drop('Status',axis=1)
y=df_cancer['Status']
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
# Check the shapes 
print(f"Training set features shape: {X_train.shape}")
print(f"Testing set features shape: {X_test.shape}")
print(f"Training set target shape: {y_train.shape}")
print(f"Testing set target shape: {y_test.shape}")
```

```python
## function  to  evaluate models

def compare_models(X_train, X_test, y_train, y_test): 
    #  models
    models = {
        'Random Forest': RandomForestClassifier(),
        'Gradient Boosting': GradientBoostingClassifier(),
        'Logistic Regression': LogisticRegression(),
        'SVC': SVC(),
        'Decision Tree': DecisionTreeClassifier(),
        'Bagging Classifier': BaggingClassifier(),
        'LightGBM': LGBMClassifier()
    } 
    
    # To save the results
    result = {}
    
    for name, model in models.items():
        # Train the model
        model.fit(X_train, y_train)
        
        # Make predictions
        prediction = model.predict(X_test)
        
        # Calculate metrics
        accuracy = accuracy_score(y_test, prediction)
        precision = precision_score(y_test, prediction, average='weighted')
        recall = recall_score(y_test, prediction, average='weighted')
        f1 = f1_score(y_test, prediction, average='weighted')
        
        # Store the metrics in the result dictionary
        result[name] = {
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1
        }
    
    # Convert the result dictionary to a DataFrame for easier viewing
    result_df = pd.DataFrame(result).T
    
    return result_df
```

```python
# use function to compare model
results_df=compare_models(X_train, X_test, y_train, y_test)
```

```python
#show result 
results_df
```

```python
# Function to plot metrics
def plot_metric(result_for_comparing, metric, palette, title):
    # Sort by  metric in descending order
    sorted_result = result_for_comparing.sort_values(by=metric, ascending=False)
    
    #  bar plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x=sorted_result.index, y=metric, data=sorted_result, palette=palette)
    plt.title(f'Model Comparison: {title}', fontsize=16)
    plt.ylabel(metric, fontsize=14)
    plt.xlabel('Model', fontsize=14)
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Plot Accuracy
plot_metric(results_df, 'Accuracy', "Blues_d", 'Accuracy')
```

```python
# Plot Precision 
plot_metric(results_df, 'Precision', "Greens_d", 'Precision')
```

```python
# Plot Recall 
plot_metric(results_df, 'Recall', "Oranges_d", 'Recall')
```

```python
# Plot F1 Score 
plot_metric(results_df, 'F1 Score', "Purples_d", 'F1 Score')
```

```python
# Function to get  best model 
def get_best_model(result_for_comparing, metric):
    # Sort by  selected metric in descending order
    sorted_result = result_for_comparing.sort_values(by=metric, ascending=False)
    # Get  name of the model with the best score for  given metric
    best_model_name = sorted_result.index[0]
    print(f"The best model based on {metric} is: {best_model_name}")
    return best_model_name
```
